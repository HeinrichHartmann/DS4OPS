# Why Python?

Pro's:
- Very mature general purpose programming language (dynamic)
- Easy to pick up
- Superbe ecosystem of libraries for everything
- Very large community
- Leading language for Data Science (2017)

Con's:
- Nativ python is slow
- No types --> Hard to build maintainable software

Alternatives:
- Excel -- Great for quick analysis of small datasets. Limited capabilities. Akward language.
- Matlab -- Proprietary. 
- R -- arguably more built-in packages. Akward programming language.
- Julia -- Like python but faster and with types. Interesting contender. Not long around.
- lua -- Fast. Minimal GP programming language. Good Library coverage (we use it a lot!)
- Java, C, C++, Go -- hard to do data exploration in compiled languages. Less library support. Great for production code.
