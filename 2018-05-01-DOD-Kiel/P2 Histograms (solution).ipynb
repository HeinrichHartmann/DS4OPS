```python
def circll_plot(H):
    x=[] # midpoints
    h=[] # height
    w=[] # widths
    for b, c in H:
        x.append(b.midpoint())
        h.append(c / b.width())
        w.append(b.width())
    plt.bar(x,h,w)
```

```python
plt.figure(figsize=(15,5),dpi=200)
circll_plot(H)
plt.xlim(0,200)
```
