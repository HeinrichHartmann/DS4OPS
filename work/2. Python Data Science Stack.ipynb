```python
B=A[:100,:5].T
```

```python
plt.plot(B.T,".-")
plt.ylim(0,100)
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
pca.fit(B)
print(pca.explained_variance_ratio_)
plt.plot(pca.components_.T)
```

```python
C = pca.transform(B)
plt.scatter(C[:,0],C[:,1])
```
