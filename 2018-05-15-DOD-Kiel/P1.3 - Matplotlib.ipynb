## Plotting with Matplotlib

In this lab you will learn how to produce the basic plot types:

1. Line Plots
2. Rugplot
3. Histograms
4. Scatterplot

with the python matplotlib library.

```python
from matplotlib import pyplot as plt
# http://matplotlib.org/
# matplotlib provides:
# - plotting primitives: figures, axes, plygons
# - advanced plotting methods for eg. histograms, scatterplots
# pyplot is a MATLAB like API to matplotlib

# Resources:
# pyplot tutorial : https://matplotlib.org/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py
# pyplot API: https://matplotlib.org/api/pyplot_summary.html
```

```python
# Configure Jupyer to show graphic output
%matplotlib inline
```

```python
# Basic Plotting Example

# Example data
X=[  1,   2,   3, 10  ]
Y=[5.3, 6.5, 8.2, 11.5]

# Step 1: Setup plotting canvas
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html#matplotlib.pyplot.figure
plt.figure(figsize=(15,5))

# Step 2: Plot!
plt.plot(X,Y)
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot
```

```python
# Step 3: Make it pretty
plt.figure(figsize=(15,5), dpi=100)
plt.plot(X,Y,"o-", linewidth=.5, markersize=3)
plt.title("A complete plot needs a title")
plt.xlim(0,12)
plt.ylim(0,15)
plt.xticks(X+[0,5,10]) # ticks are the little indicators at the axis
plt.yticks(Y+[0,1,2,3,15])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot([1,10],[5.3,11.5],"r--", alpha=0.5) # draw a second plot onto the same canvas!
plt.legend(["This line is really important line","direct way"], loc='upper right', frameon=False)

# Step 4 (Optional): Export the plot as image
plt.savefig("example.png")
```
